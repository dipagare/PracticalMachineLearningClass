```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('titanic_train.csv', header=0)
test_df = pd.read_csv('titanic_test.csv', header=0)
```

```python
train_df.head()
```

EXPLORACION DE VARIABLES

```python
# check for missing values
titanic = train_df
titanic.isnull().sum()
```

```python
titanic.loc[titanic.Embarked.isnull(), 'Embarked'] = titanic.Embarked.mode().values
titanic['Sex_Female'] = (titanic['Sex'] == 'female').astype(np.int)
embarked_dummies = pd.get_dummies(titanic.Embarked, prefix='Embarked')
embarked_dummies.drop(embarked_dummies.columns[0], axis=1, inplace=True)
titanic = pd.concat([titanic, embarked_dummies], axis=1)
```

```python
titanic.ix[(titanic.Age<0),"condicion"]=0 # vacios
titanic.ix[(titanic.Age>=0),"condicion"]=1 #Age
```

```python
titanic.isnull().sum()
```

```python
data_filter = titanic[titanic['condicion'] == 1]
data_filter.head(200)
len(data_filter)
```

```python

from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
feature_cols = ['Survived','Pclass', 'Parch', 'Sex_Female', 'Embarked_Q', 'Embarked_S']
X = data_filter[feature_cols]
y =data_filter.Age
linreg.fit(X, y)
```

```python
data_vacios = titanic[titanic['condicion'] != 1]
from sklearn.linear_model import LinearRegression
data_vacios.head()
X_vacios = data_vacios[feature_cols]
data_vacios['Age'] = linreg.predict(X_vacios)
data_vacios.head()
```

```python
 completo = pd.DataFrame(data_filter)
    
imputados = pd.DataFrame (data_vacios)

final = completo.append(imputados)
final.shape
```

```python
# define X and y

feature_cols = ['Pclass', 'Parch', 'Age','Age_Cuadro','Sex_Female', 'Embarked_Q', 'Embarked_S','Fare_Cuadro','Persona_Dama_Madura',
               'Persona_Hombre_Joven','Persona_Hombre_Maduro','Pclass_+1','Age_Fare']
X = final[feature_cols]
y = final.Survived

# train/test split
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# train a logistic regression model
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e9)
logreg.fit(X_train, y_train)

# make predictions for testing set
y_pred_class = logreg.predict(X_test)

# calculate testing accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred_class))
```

USANDO CROSS_VALIDACION

```python
# Create k-folds
from sklearn.cross_validation import KFold
kf = KFold(X.shape[0], n_folds=15, random_state=0)

results = []
models = []
for train_index, test_index in kf:
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # train a logistic regression model
    models.append(LogisticRegression(C=1e9))
    
    models[-1].fit(X_train, y_train)

    # make predictions for testing set
    y_pred_class = models[-1].predict(X_test)

    # calculate testing accuracy
    results.append(metrics.accuracy_score(y_test, y_pred_class))
```
